Task 1 (Variant 1).

```python
dist = 10
summ = dist
r = 10
for i in range(2, 8):
  x = (dist * r) / 100
  dist = dist + x
  summ = summ + dist
print(summ)
```

Task 1 (Variant 2).

```python
dist = 10
summ = dist
r = 10
k = 2
while k <= 7:
  x = (dist * r) / 100
  dist = dist + x
  summ = summ + dist
  k = k + 1
print(summ)
```
